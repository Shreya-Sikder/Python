```python
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

Genarate Datasets

```python
n_samples= 100000
n_features=2
n_cluster= 4
```

```python
X, _ = make_blobs(n_samples= n_samples, centers=n_cluster, n_features=n_features, random_state=42)
```

```python
X.shape
```

```python
kmeans = KMeans(n_clusters=n_cluster, random_state=42, n_init=10, max_iter=300)
```

```python
kmeans.fit(X)
```

```python
centers =kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
print(centers)
```

```python
print(labels)
```

Visualization


```python
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1], c=lables ,cmap="viridis", alpha=0.6)
plt.title("K-Means Clustering")
plt.xlabel("Feature  1")
plt.ylabel("Feature 2")
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X[:, 0], X[:, 1], c= labels, cmap="viridis", alpha=0.6)
plt.scatter(centers[:, 0], centers[:, 1], c="red", marker='x', s=200, label="centers")
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
